```python
variable = 'Dye [concentration]'

import textwrap
import matplotlib.pyplot as plt
import seaborn as sns
def plot_simple_comparisons(df,x,hue,y='Percent Replicating',order=None,hue_order=None,
col=None, col_order=None, col_wrap=None,row=None,row_order=None,jitter=0.5,dodge=True,plotname=None,
sharex=True):
    plt.rcParams["legend.markerscale"] =1.5
    sns.set_style("ticks")
    sns.set_context("paper",font_scale=1.5)
    g = sns.catplot(data=df, x = x ,y = y, order=order,
    hue=hue, hue_order=hue_order, col=col, col_order = col_order, col_wrap=col_wrap,row=row,
    row_order = row_order, palette='Set1',s=12,linewidth=1,jitter=jitter,
    alpha=0.8,dodge=dodge,sharex=sharex,aspect=1.4)
    g.set(ylim=([40,80]))
    g.figure.subplots_adjust(wspace=0)
    for _,ax in g.axes_dict.items():
        labels_out = []
        labels=ax.get_xticklabels()
        for label in labels:
            labels_out.append(textwrap.fill(str(label)[12:-2], width=11,break_long_words=False))
        ax.set_xticklabels(labels=labels_out,rotation=0)
    g.figure.subplots_adjust(wspace=0.1,hspace=0.25)
    if plotname:
        plotname = f"../figures/{plotname}"
    else:
        plotname = f"../figures/{x}-{hue}.png"
    g.savefig(plotname,dpi=300)
    print(f'Saved to {plotname}')

plot_simple_comparisons(df=plate_df_individual,x=variable,hue='Batch',hue_order=['Stain2','Stain3','Stain4','Stain5'],
col='Dye',col_order = dyes_perturbed,col_wrap=4,sharex=False)
```

```python
variable = 'Dye [concentration]'
dyes_perturbed = ['DNA','ER','Actin','Mito','ER & RNA','ER & Actin','Actin & Golgi-Plasma Membrane']


import textwrap
import matplotlib.pyplot as plt
import seaborn as sns
def plot_simple_comparisons(df,x,hue,y='Percent Replicating',order=None,hue_order=None,
col=None, col_order=None, col_wrap=None,row=None,row_order=None,jitter=0.5,dodge=True,plotname=None,
sharex=True):
    plt.rcParams["legend.markerscale"] =1.5
    sns.set_style("ticks")
    sns.set_context("paper",font_scale=1.5)
    g = sns.catplot(data=df, x = x ,y = y, order=order,
    hue=hue, hue_order=hue_order, col=col, col_order = col_order, col_wrap=col_wrap,row=row,
    row_order = row_order, palette='Set1',s=80,linewidth=1,jitter=jitter,
    alpha=0.8,dodge=dodge,sharex=sharex,aspect=1.4)
    g.set(ylim=([40,80]))
    g.figure.subplots_adjust(wspace=0)
    for _,ax in g.axes_dict.items():
        labels_out = []
        labels=ax.get_xticklabels()
        for label in labels:
            labels_out.append(textwrap.fill(str(label)[12:-2], width=11,break_long_words=False))
        ax.set_xticklabels(labels=labels_out,rotation=0)
    g.figure.subplots_adjust(wspace=0.1,hspace=0.25)
    if plotname:
        plotname = f"../figures/{plotname}"
    else:
        plotname = f"../figures/{x}-{hue}.svg"
    g.savefig(plotname,dpi=300)
    print(f'Saved to {plotname}')

plot_simple_comparisons(df=plate_df_individual,x=variable,hue='Batch',hue_order=['Stain2','Stain3','Stain4','Stain5'],
col='Dye',col_order = dyes_perturbed,col_wrap=4,sharex=False)
```
