```python
import os
import pandas
pandas.__version__
```

```python
batchlist = os.listdir('../profiles')
batchlist = [batch for batch in batchlist if '.' not in batch]
```

```python
feature_num_dict = {}
selected_feature_num_dict = {}
plate_len_dict = {}
batch_count = 0
total_plate_count = 0
for eachbatch in batchlist:
    count = 0
    platelist = os.listdir(os.path.join('..','profiles',eachbatch))
    batch_feature_num_dict = {}
    selected_batch_feature_num_dict = {}
    for eachplate in platelist:
        fullplatename = os.path.join('..','profiles',eachbatch,eachplate,eachplate+'.csv.gz')
        if os.path.exists(fullplatename):
            count +=1
            feature_len = pandas.read_csv(fullplatename).shape[1]
            if feature_len not in batch_feature_num_dict.keys():
                batch_feature_num_dict[feature_len] = [eachplate]
            else:
                batch_feature_num_dict[feature_len] += [eachplate]
            if os.path.exists(os.path.join('..','profiles',eachbatch,eachplate,eachplate+'_normalized_feature_select_dmso.csv.gz')):
                fullselectedplatename = os.path.join('..','profiles',eachbatch,eachplate,eachplate+'_normalized_feature_select_dmso.csv.gz')
            elif os.path.exists(os.path.join('..','profiles',eachbatch,eachplate,eachplate+'_normalized_feature_select_negcon.csv.gz')):
                fullselectedplatename = os.path.join('..','profiles',eachbatch,eachplate,eachplate+'_normalized_feature_select_negcon.csv.gz')
            else:
                fullselectedplatename = False
            if fullselectedplatename:
                feature_len = pandas.read_csv(fullselectedplatename).shape[1]
                if feature_len not in selected_batch_feature_num_dict.keys():
                    selected_batch_feature_num_dict[feature_len] = [eachplate]
                else:
                    selected_batch_feature_num_dict[feature_len] += [eachplate]  
    feature_num_dict[eachbatch] = batch_feature_num_dict
    selected_feature_num_dict[eachbatch] = selected_batch_feature_num_dict
    plate_len_dict[eachbatch] = count
    batch_count += 1
    total_plate_count += count
```

```python
print(f'In total, {batch_count} batches with {total_plate_count} plates were assessed')
```

```python
batchlist.sort()
batch_level_selected_list = []
for eachbatch in batchlist:
    platecount = plate_len_dict[eachbatch]
    raw_feat_num = len(list(feature_num_dict[eachbatch].keys()))
    selected_feat_num = len(list(selected_feature_num_dict[eachbatch].keys()))
    if selected_feat_num == 1: 
        if platecount != 1:
            batch_level_selected_list.append(eachbatch)
    print(f"Batch {eachbatch} has {platecount} plates, {raw_feat_num} lengths of raw features, and {selected_feat_num} lengths of selected features")
```

```python
if len(batch_level_selected_list) == 0:
    print('No batches are suspected to have undergone batch level selection')
else:
    print(f'{len(batch_level_selected_list)} batch(es) is/are suspected to have undergone batch level selection')
    for eachbatch in batch_level_selected_list:
        print(eachbatch)
```
