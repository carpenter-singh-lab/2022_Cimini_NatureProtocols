```python
import pandas

from utils import calculate_percent_replicating_Target, plot_simple_comparison
```

```python
variable = 'Fields of view'
variable_to_save = variable.replace(' ','_')

plate_df = pandas.DataFrame()

for eachfov in range(1,16):
    for eachsubsample in ['a','b','c']:

        plate_df = plate_df.append({'Plate Name List' : ['BR00117015','BR00117016','BR00117017','BR00117019'], 
        variable : eachfov, 'Batch' : f'CPJUMP1', 'Percent Replicating' : calculate_percent_replicating_Target(
            f'../profiles-cp-jump1-fov/profiles/2020_11_04_CPJUMP1_{eachfov}{eachsubsample}',['BR00117015','BR00117016','BR00117017','BR00117019'],
            suffix='_normalized_feature_select_negcon_plate.csv.gz'),}, ignore_index=True)


plate_df = plate_df.append({'Plate Name List' : ['BR00117015','BR00117016','BR00117017','BR00117019'], 
variable : 16, 'Batch' : f'CPJUMP1', 'Percent Replicating' : calculate_percent_replicating_Target(
    f'../profiles-cp-jump1-fov/profiles/2020_11_04_CPJUMP1_16',['BR00117015','BR00117016','BR00117017','BR00117019'],
     suffix='_normalized_feature_select_negcon_plate.csv.gz'),}, ignore_index=True)
```

```python
#You may find that running this a couple of times is needed to find a version where the jitter is about right
plot_simple_comparison(df=plate_df,x=variable,y='Percent Replicating',hue='Batch',ylim=[50,100])
```

```python
plate_df.to_csv(f'checkpoint_csvs/{variable_to_save}-replicating.csv',index=False)
```
